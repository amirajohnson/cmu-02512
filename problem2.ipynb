```python
import numpy as np 
import matplotlib.pyplot as plt 
from pprint import pprint
```

```python
#Borrowed from website bc the point is not to code it.
theta = 1000
c = 0.01
a = 1
epsilon = .0005
gamma = 0.01
```

```python
n1, n2 = np.meshgrid(np.linspace(1, 1000, 30), np.linspace(1, 1000, 30)) 

# Directional vectors 
u = (theta - a*c*n1*n2)  #n1 change
v = (epsilon*a*c*n1*n2 - gamma*n2) #n2 change

n2_null, n1_null = theta / (a * c * n1), gamma / epsilon*a*c*n2


# Plotting Vector Field with QUIVER 
plt.quiver(n1, n2, u, v, color = 'b')
plt.xlabel('n1')
plt.ylabel('n2')

#calculating the nullclines
nullcline1 = np.linspace(0, 1000, 10)
f1_nullcline2 = theta/(a*c*nullcline1)
f2_nullcline2 = (epsilon*a*c*nullcline1)/gamma

plt.plot(nullcline1, f1_nullcline2, label = "Nullcline 1")
plt.plot(nullcline1, f2_nullcline2, label = "Nullcline 2")



plt.title("Vector Field and Nullclines") 

plt.legend()
plt.grid() 
plt.show() 


```
